```python
# Importing required libraries
import pandas as pd
import nltk
import string
import re
from sklearn.metrics import classification_report
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import GridSearchCV
```

# Task 1

```python
# Reading the csv and ignoring the lines causing errors
df = pd.read_csv('/Users/mahvashmaghrabi/Downloads/amazon_reviews_us_Mobile_Electronics_v1_00.tsv', sep='\t', error_bad_lines=False)
df
```

# Text Preprocessing

```python
# Preprocessing the text
def preprocess_text(text):

# Tokenize
    text = str(text) 
    tokens = word_tokenize(text.lower())
    
# Removing stop words

    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]

# Lemmatization

    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]


# Joining the tokens

    processed_text = ' '.join(lemmatized_tokens)

    return processed_text

# Applying the function df

df['review_headline'] = df['review_headline'].apply(preprocess_text)
df['review_body'] = df['review_body'].apply(preprocess_text)
df
```

```python
# Remove punctuations
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return str(text).translate(translator)

df['review_headline'] = df['review_headline'].apply(lambda x: remove_punctuation(x))
df['review_body'] = df['review_body'].apply(lambda x: remove_punctuation(x))

# Printing to check the difference in the reviews
print(df['review_headline'])
print(df['review_body'])
```

```python
# Removing numbers and float values from the reviews
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

df['review_body'] = df['review_body'].apply(remove_numbers)
df['review_headline'] = df['review_headline'].apply(remove_numbers)
```

```python
# Printing the star rating column
print(df['star_rating'])
```

```python
# Defining a function to map star ratings to sentiment labels
def map_star_to_sentiment(star_rating):
    if star_rating <= 2:
        return 0
    else:
        return 1

# Creating a new column "sentiment_label"
df["star_rating_label"] = df["star_rating"].map(map_star_to_sentiment)
```

# Sentiment Analysis

```python
# Initializing the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Loop through each text in the "review_body" column 
labels = []
for text in df["review_body"]:
    scores = sia.polarity_scores(text)
    if scores['compound'] > 0:
        labels.append("positive")
    elif scores['compound'] < 0:
        labels.append("negative")
    else:
        labels.append("neutral")

# Add the labels as a new column "sentiment"
df["sentiment"] = labels

# Printing both the reviews and the corresponding sentiment
print(df[[ "review_body","sentiment"]])
```

```python
# Converting sentiment analysis results to numerical values
def convert_sentiment(sentiment_string):
    if sentiment_string == "positive":
        return 1
    else:
        return 0

# Applying the function
df["sentiment_numerical"] = df["sentiment"].apply(convert_sentiment)

# Printing the new column
print(df["sentiment_numerical"])
```

```python
# Printing the first 50 rows
df.head(50)
```

```python
# Printing the classification report to check the accuracy of the sentiment analysis results
print(classification_report(df['star_rating_label'], df['sentiment_numerical']))
```
